```python
from num2words import num2words
invIndex={}
length={}
df={}
global_doc=[]
templist=[]
gd_list={}

def preprocessing():
    counter=0
    import glob
    import re
    import string
    from nltk.corpus import stopwords
    from nltk.tokenize import word_tokenize
    from nltk.stem import WordNetLemmatizer
    folders = [i for i in glob.glob('20_newsgroups/*')]
    for j in folders:
        print(j)
        subfolders=glob.glob(j+'/*')
        for h in subfolders:
            
            templist=[]
            file=open(h)
            
            X=file.read()
            counter=counter+1
            global_doc.append(counter)
            
            X =X.lower()
            
           
            
            
            translator=X.maketrans('','',string.punctuation)
            #translator=X.maketrans(string.punctuation,'                                ')
            
            X=X.translate(translator)
            #print(X)
            stop_words = set(stopwords.words('english'))
            tokens = word_tokenize(X)
            
            #tokens=list(set(tokens))

#             pattern='[0-9][a-z]|[0-9]'
#             tokens=[re.sub(pattern,'',j) for j in tokens]
            
            result = [i for i in tokens if not i in stop_words]
    #print (result)

            lemmatizer=WordNetLemmatizer()
            for word in result:
                word1=lemmatizer.lemmatize(word)
                if word1.isdecimal():
                    templist.append(num2words(word1))
                else:
                    templist.append(word1)
            templist1=list(set(templist))
            #print(templist[2])
            for w in templist1:
                if w in df.keys():
                    df[w]=df[w]+1
                else:
                    df[w]=1
                    
                if w in invIndex.keys():
                    invIndex[w][counter]=templist.count(w)
                    gd_list[w][counter]=gd[str(counter)]
                    
                else:
                    invIndex[w]={}
                    invIndex[w][counter]=templist.count(w)
                    gd_list[w]={}
                    gd_list[w][counter]=gd[str(counter)]


            length[counter]=len(templist)        


preprocessing() 
print("pre processing completed")
#print(invIndex[2])
#print(global_doc)
```

```python
import operator
for h in invIndex.keys():
    invIndex[h]=dict(sorted(invIndex[h].items(),key=operator.itemgetter(1),reverse=True))
    gd_list[h]=dict(sorted(gd_list[h].items(),key=operator.itemgetter(1),reverse=True))
print(invIndex["gnuplot"])
     
```

```python
high={}
low={}
r=20
for term in invIndex:
    num=r
    high[term]={}
    low[term]={}
    for docs in invIndex[term]:
        if num>=0:
            high[term][docs]=invIndex[term][docs]
            num-=1
        else:
            low[term][docs]=invIndex[term][docs]
            
print(high["gnuplot"])
print(low["gnuplot"])
```

```python
import string 
import re
import string
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
query=input()
print("Number of documents to retrive")
k=int(input())

query_list=[]

#query_list ="The three little pigs started to feel they need a real home"
  
query=query.lower()
#translator=query.maketrans(string.punctuation,'                                ')
query=query.translate(str.maketrans("","",string.punctuation))
#query=query.translate(translator)
stop_words = set(stopwords.words('english'))
query = word_tokenize(query)
query=list(query)
result = [i for i in query if not i in stop_words]
querylist=[]
lemmatizer=WordNetLemmatizer()
for word in result:
    if word.isdecimal()==True:
        word=num2words(word)
    querylist.append(lemmatizer.lemmatize(word))
print(querylist)

```

```python
dict_tfidf={}
for term in querylist:
    if term in invIndex.keys():
        for doc2 in high[term]:
            if doc2 not in dict_tfidf:
                dict_tfidf[doc2]=high[term][doc2]*idf[term]
            else:
                dict_tfidf[doc2] +=high[term][doc2]*idf[term]

#print(dict_tfidf[19723])
print(len(dict_tfidf)) 


if k>(len(dict_tfidf)):
    
        for term in querylist:
            if term in invIndex.keys():
                for doc4 in low[term]:
                    if doc4 not in dict_tfidf:
                        dict_tfidf[doc4]=low[term][doc4]*idf[term]
                    else:
                        dict_tfidf[doc4] +=low[term][doc4]*idf[term]

#print(len(dict_tfidf))
for doc3 in dict_tfidf:
    dict_tfidf[doc3]=dict_tfidf[doc3]/length[doc3]+gd[str(doc3)]/maxgd
    
# dict_tfidf=dict(sorted(dict_tfidf.items(),key=operator.itemgetter(1),reverse=True))
# print(dict_tfidf)

```

```python
def rank(k):
    from heapq import nlargest
    kHighest = nlargest(k,dict_tfidf,key = dict_tfidf.get)
    
    
    for val in kHighest: 
        print(val, ":",dict_tfidf.get(val))
rank(k)        
```
